Importing all Dependencies

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

Data Collection and Processing

```python
#Loading the dataset to a pandas DataFrame
parkinsons_data = pd.read_csv(r"C:\Users\User\ML_crush_course\ML_challenges\parkinsons_data.csv")
```

Labeling the data

```python
X = parkinsons_data.drop(columns=['name', 'status'], axis=1)

Y = parkinsons_data['status']
```

Standardazing data

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
scaler.transform(X)
```

```python
print(X)
```

Splitting data into training and test data

```python
X_train, X_test, Y_train, Y_test= train_test_split(
    X, 
    Y,
    train_size =0.2,
    stratify=Y,
    random_state=2
)
```

```python
print(X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
```

Training data

```python
model = SVC(kernel='linear')
```

```python
model.fit(X_train, Y_train)
```

Evaluating the data

```python
#Checking the accuracy score on the training data
trainin_prediction_data = model.predict(X_train)

accuracy_trainin_data = accuracy_score(trainin_prediction_data, Y_train)

print("Accuracy score of the training data: ", accuracy_trainin_data)
```

```python
#Checking the accuracy score on the test data

test_prediction_data = model.predict(X_test)

accuracy_test_data = accuracy_score(test_prediction_data, Y_test)

print("Accuracy score of the test data: ", accuracy_test_data)
```
